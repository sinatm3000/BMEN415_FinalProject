```python
import numpy as np  # Importing NumPy library
import pandas as pd  # Importing Pandas library
import matplotlib.pyplot as plt  # Importing Matplotlib library's "pyplot" module
import seaborn as sns  # Importing Seaborn library

import os
```

```python
data = pd.read_csv("Breast Cancer Wisconsin.csv")
```

```python
fig, axes = plt.subplots(figsize=(8, 8)) 
sns.heatmap(data=data.corr(), linewidths=.75, ax=axes) 
plt.show()
```

```python
X = data.drop(["id","diagnosis"], axis=1) 
y = data.diagnosis.values
```

```python
data['diagnosis'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)
```

```python
#from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=55)
rf_model.fit(x_train, y_train)
```

```python
from sklearn import metrics

print("Training Score: ")
print(rf_model.score(x_train, y_train))
print("Testing Score:")
print(rf_model.score(x_test, y_test))
```

```python
accuracy_train = []
accuracy_test =[]

for i in range (1,50):
    rf_model = RandomForestClassifier(random_state = 55, max_depth=i)
    rf_model.fit(x_train, y_train)
    accuracy_train.append(rf_model.score(x_train,y_train))
    accuracy_test.append(rf_model.score(x_test,y_test))

plt.figure(figsize=(5,5))
plt.plot(accuracy_train, color="blue", marker="o", markersize=5, label="Training Accuracy")
plt.plot(accuracy_test, color="red", marker="o", markersize=5, label="Testing Accuracy")
plt.title("Accuracy vs. Max Tree Depth")
plt.xlabel("Max Tree Depth")
plt.ylabel("Model Accuracy")
plt.legend()
plt.show()

print("Maximum Accuracy: ", max(accuracy_test),"at Max Depth =", accuracy_test.index(max(accuracy_test)))
```

```python
rf_model = RandomForestClassifier(random_state=55, max_depth=2)
rf_model.fit(x_train, y_train)

print("Training Score: ")
print(rf_model.score(x_train, y_train))
print("Testing Score:")
print(rf_model.score(x_test, y_test))

```

```python
from sklearn.model_selection import cross_val_predict  # For K-Fold Cross Validation
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

y_pred_train = cross_val_predict(rf_model, x_train, y_train, cv=5)
y_pred_test = cross_val_predict(rf_model, x_test, y_test, cv=5)

cm = confusion_matrix(y_test, y_pred_test)
accuracy_score(y_test, y_pred_test)

sns.heatmap(cm, annot=True)
```
