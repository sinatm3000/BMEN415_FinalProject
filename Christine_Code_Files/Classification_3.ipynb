```python
X = data.drop(["id","diagnosis"], axis=1) 
y = data.diagnosis.values
```

```python
data['diagnosis'].value_counts()
```

```python
#Features are highly correlated so feature reduction is required
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Scale data before applying PCA
scaling=StandardScaler()
 
# Use fit and transform method
scaling.fit(X)
Scaled_data=scaling.transform(X)
 
# Set the n_components=3
principal=PCA(n_components=3)
principal.fit(Scaled_data)
x_revised=principal.transform(Scaled_data)
 
# Check the dimensions of data after PCA
print(x_revised.shape)
#from sklearn.preprocessing import StandardScaler

#Scaled_x_train = StandardScaler().fit(x_train)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_revised, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
Log_Regression = LogisticRegression()
Log_Regression.fit(x_train, y_train)
```

```python
from sklearn import metrics
print("Training Score: ")
print(Log_Regression.score(x_train, y_train))
print("Testing Score:")
print(Log_Regression.score(x_test, y_test))
```
