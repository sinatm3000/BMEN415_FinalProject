```python
data.drop(["CAS", "SMILES", "Set", "Class"], axis=1, inplace=True)
```

```python
data.info()
```

```python
X = data.drop(["logBCF"], axis=1) 
y = data.logBCF.values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)
```

```python
#Scaling Data
from sklearn import preprocessing
scaling = preprocessing.StandardScaler()

y_train=np.reshape(y_train, (-1,1))
y_test=np.reshape(y_test, (-1,1))

scaling.fit(x_train)
Scaled_x_train = scaling.transform(x_train)

scaling.fit(x_test)
Scaled_x_test = scaling.transform(x_test)

scaling.fit(y_train)
Scaled_y_train = scaling.transform(y_train)

scaling.fit(y_test)
Scaled_y_test = scaling.transform(y_test)

Scaled_y_train = np.reshape(Scaled_y_train, -1)
Scaled_y_test = np.reshape(Scaled_y_test, -1)
```

```python
from sklearn.svm import SVR

Support_Vector = SVR(kernel='rbf', C=2, epsilon=0.1)
Support_Vector.fit(Scaled_x_train, Scaled_y_train)
```

```python
from sklearn.model_selection import cross_val_predict  # For K-Fold Cross Validation
from sklearn.metrics import r2_score  # For find accuracy with R2 Score
from sklearn.metrics import mean_squared_error  # For MSE
from math import sqrt  # For squareroot operation

y_pred_train = cross_val_predict(Support_Vector, Scaled_x_train,Scaled_y_train,cv=5)
y_pred_test = cross_val_predict(Support_Vector, Scaled_x_test, Scaled_y_test, cv=5)

accuracy_train = r2_score(Scaled_y_train, y_pred_train)
print("Training Accuracy for Suport Vector Machine Regression Model: ", accuracy_train)

accuracy_test = r2_score(Scaled_y_test, y_pred_test)
print("Testing Accuracy for Support Vector Machine Regression Model: ", accuracy_test)

RMSE_train = sqrt(mean_squared_error(Scaled_y_train, y_pred_train))
print("RMSE for Training Data: ", RMSE_train)

RMSE_test = sqrt(mean_squared_error(Scaled_y_test, y_pred_test))
print("RMSE for Testing Data: ", RMSE_test)


```

```python
#To find the optimal amount of regularization, we will iterate over the values of 'C' and plot the training and testing scores 
accuracy_train = []
accuracy_test =[]

for i in range (1,10):
    Support_Vector = SVR(kernel='rbf', C= i, epsilon=0.1)
    Support_Vector.fit(Scaled_x_train, Scaled_y_train)
    y_predict_train = cross_val_predict(Support_Vector, Scaled_x_train, Scaled_y_train,cv=5)
    y_predict_test = cross_val_predict(Support_Vector, Scaled_x_test, Scaled_y_test, cv=5)
    accuracy_train.append(r2_score(Scaled_y_train, y_predict_train))
    accuracy_test.append(r2_score(Scaled_y_test,y_predict_test))

plt.figure(figsize=(5,5))
plt.plot(accuracy_train, color="blue", marker="o", markersize=5, label="Training Accuracy")
plt.plot(accuracy_test, color="red", marker="o", markersize=5, label="Testing Accuracy")
plt.title("Accuracy vs. Model Complexity")
plt.xlabel("C Value")
plt.ylabel("Model Accuracy")
plt.legend()
plt.show()

print("Maximum Accuracy: ", max(accuracy_test),"at C Value =", accuracy_test.index(max(accuracy_test)))
```

```python
true_value_train = Scaled_y_train
predicted_value_train = y_pred_train

plt.figure(figsize=(5,5))
p1 = max(max(predicted_value_train), max(true_value_train))
p2 = min(min(predicted_value_train), min(true_value_train))
plt.plot([p1,p2], [p1,p2], '-', c="black")
plt.scatter(true_value_train, predicted_value_train, color="blue")
plt.title("SVM (Rbf Kernel): True vs. Predicted Values (Training Set)")
plt.xlabel("True")
plt.ylabel("Predicted")
plt.show()
```

```python
true_value_test = Scaled_y_test
predicted_value_test = y_pred_test

plt.figure(figsize=(5,5))
q1 = max(max(predicted_value_test), max(true_value_test))
q2 = min(min(predicted_value_test), min(true_value_test))
plt.plot([q1,q2], [q1,q2], '-', c="black")
plt.scatter(true_value_test, predicted_value_test, color="blue")
plt.title("SVM (Rbf Kernel): True vs. Predicted Values (Testing Set)")
plt.xlabel("True")
plt.ylabel("Predicted")
plt.show()
```
